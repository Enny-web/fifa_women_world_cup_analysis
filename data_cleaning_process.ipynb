```python
import pandas as pd
import numpy as np
import enchant   #for checking spelling
```

```python
#Let's declare the null values

null_values = ['na','N/a','NaN',np.nan]
```

```python
#Now let's import the data
```

```python
data = pd.read_csv('datathon data.csv',na_values =null_values)
```

```python
data.head(5)     #Checks the first 5 rows of the data
```

```python
data.tail(5)     #Checks the last 5 rows of the data.
```

```python
#Checking for duplicates

data.duplicated().sum()
```

The Data Contains no duplicate values

```python
#Data has Unnamed colums,let's get rid of them


data = data.loc[:, ~data.columns.str.contains('^Unnamed')]
```

```python
data.head(2)
```

Unnamed columns have been removed

```python
#Checking for null values

data.isnull().sum()
```

The data has 40 null values in the possesion column and 74 null values each in the yellow cards and red cards columns respectively

```python
# But first possesion is a wrong spelling ,now Let's correct it!

data.rename(columns={'possesion': 'possession'}, inplace=True)
```

```python
data.columns
```

The spelling error has been rectified

# Filling up the null values;

We can fill the yellow cards and red cards columns with zero but the possession value can not be zero so we fill up with mean of the total possesssions



```python
data['yellow_cards'].fillna(0, inplace=True)
```

```python
data['red_cards'].fillna(0, inplace=True)
```

```python
#For the possession column,let's declare a value for mean

mean_possession = data['possession'].mean()
```

```python
data['possession'].fillna(mean_possession, inplace=True)
```

```python
data.isnull().sum()
```

No null values present

Now, Let's check for spelling error

```python
import pandas as pd
import enchant

# create an enchant dictionary
d = enchant.Dict("en_US")

# loop through all string columns and detect misspelled words
for col in data.select_dtypes(include=['object']).columns:
    for i, val in enumerate(data[col]):
        words =val
        for word in words:
            if not d.check(word):
                print(f"Misspelled word '{word}' found in row {i} of column '{col}'")
```

A spelling error is detected in row 31

```python
print(data.iloc[31])
```

At row 31, cote d'ivoire has been spelt incorrectly

```python
# Correcting this spelling

data.loc[31, 'squad'] = 'Cote d\'Ivoire'
```

```python
data.iloc[31]
```

Spelling error is now corrected!

```python
data.head(3)
```

 The age column is in the 'float' format, this is very incorrect and should be converted to 'int'(integer)

```python
#Converting to int

data['age'] = data['age'].astype(int)
```

```python
data.head(3)
```

```python
#Let's do the same for the minutes played-90s column to make the data look even cleaner


data['minutes_played_90s'] = data['minutes_played_90s'].astype(int)
```

```python
data.head(2)
```

```python
#Now for yellow cards,red cards and goals plus assists per 90

data['yellow_cards'] = data['yellow_cards'].astype(int)
data['red_cards'] = data['red_cards'].astype(int)
data['goals_plus_assists_per_90'] = data['goals_plus_assists_per_90'].astype(int)
```

```python
data.head(2)
```
