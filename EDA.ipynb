###### Let's take a closer look at this data. Classifying each squad into their respective continents can help us drive further insights

```python
# To do this, let's import another library called pycountry_convert


import pycountry_convert as pc
```

```python
def get_continent(country):
    try:
        country_code = pc.country_name_to_country_alpha2(country, cn_name_format="default")
        continent_code = pc.country_alpha2_to_continent_code(country_code)
        continent_name = pc.convert_continent_code_to_continent_name(continent_code)
        return continent_name
    except:
        return np.nan

data['Continent'] = data['squad'].apply(get_continent)
```

```python
data.head(2)   #The Continent column has been added.
```

Now, Let's see the continent with the highest number of goals

```python
pivot_table2 = pd.pivot_table(data, values='goals', index='Continent', columns='year',aggfunc ='sum')

pivot1 = pivot_table2.assign(total=pivot_table2.sum(axis=1))
```

```python
pivot1
```

```python

sort1 = pivot1.sort_values(by='total', ascending=False)
```

```python
sort1
```

###### The continent with the highest number of goals is Europe and the least number of goals is scored by the squads from the African continent
